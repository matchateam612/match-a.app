```python
import requests
import pandas as pd
from io import StringIO
import time
```

```python
url = 'https://studentorg.ucsd.edu/'
response = requests.get(url)
print(f"Status Code: {response.status_code}")
html_string = StringIO(response.text)
all_org_df = pd.read_html(html_string)[0]
```

```python
text_splits = response.text.split("<a href=")
```

```python
def president_info(scrape_text):
    org_url = scrape_text[1:20]
    response = requests.get("https://studentorg.ucsd.edu/" + org_url)
    html_string = StringIO(response.text)
    df = pd.read_html(html_string)
    president_name = list(df[0].iloc[0])[0]
    president_email = list(df[0].iloc[0])[1]
    return president_name, president_email
```

```python
org_info.to_csv("org_info")
```

```python
datalist = []
for i in range(700,751):
    org_name = all_org_df.iloc[i,0]
    pres_name, pres_mail = president_info(text_splits[i+6])
    datalist.append([org_name, pres_name, pres_mail])
    print(i, "/ 300")
df = pd.DataFrame(datalist)
df.columns = ["org_name", "pres_name", "pres_mail"]
org_info = pd.concat([org_info, df], ignore_index=True)
```
